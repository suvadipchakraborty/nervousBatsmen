```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/suvad/Downloads/323_7768_bundle_archive/deliveries.csv")
```

```python
df.head(10)
```

```python
df = df[df['inning'] <= 2]
```

```python
df = df.sort_values(["match_id", "inning", "batsman", "over","ball"], ascending = (True, True, True, True, True))
```

```python
df.head(50)
```

```python
df1 = df
df1['sum_runs'] = df1.groupby(by=["match_id", "inning", "batsman"])['batsman_runs'].cumsum()
df1['sum_balls'] = df1.groupby(by=["match_id", "inning", "batsman"])['match_id'].cumcount()
df['run_tens'] = (df['sum_runs'] / 10).astype(int) *10
df['match_ball'] = ((df['over']-1)*6) + df['ball']
```

```python
df1.head(50)
```

```python
df1['idx'] = pd.Categorical(df1['match_id'].astype(str) + '_' + df['batsman'].astype(str))
```

```python
df1.head(50)
```

```python
df_max = pd.DataFrame(df1.groupby('idx')["sum_runs"].max()).reset_index()
```

```python
df_max
```

```python
df_max_50_plus = df_max[df_max['sum_runs'] > 50]
```

```python
df_max_50_plus
```

```python
df_max_50_plus_excl = df1[df1['run_tens'] == 40]
df_max_50_plus_excl = pd.DataFrame(df_max_50_plus_excl.groupby('idx')["match_ball"].min()).reset_index()
```

```python
df_max_50_plus_excl
```

```python
df2 = df1[df1.idx.isin(df_max_50_plus.idx)]
```

```python
df2
```

```python
df2 = pd.merge(left=df2, right=df_max_50_plus_excl, how='left', on='idx')
```

```python
df2
```

```python
df2['idx_tens'] = pd.Categorical(df2['idx'].astype(str) + '_' + df2['run_tens'].astype(str))
df2["start_ball"] = df2.groupby(['idx_tens'])["match_ball_x"].transform(min)
df2["end_ball"] = df2.groupby(['idx_tens'])["match_ball_x"].transform(max)
df2["balls_survived_tens"] = (df2["end_ball"] - df2["start_ball"]) + 1
df2["balls_faced_tens"] = df2.groupby(['idx_tens'])["match_ball_x"].transform('count')
```

```python
df2
```

```python
df2.to_csv('nervous.csv', encoding='utf-8', index=False)
```

```python
df_max_100_plus_excl = df1[df1['run_tens'] == 90]
df_max_100_plus_excl = pd.DataFrame(df_max_100_plus_excl.groupby('idx')["match_ball"].min()).reset_index()
```

```python
df_max_100_plus = df_max[df_max['sum_runs'] > 100]
df3 = df1[df1.idx.isin(df_max_100_plus.idx)]
df3 = pd.merge(left=df3, right=df_max_100_plus_excl, how='left', on='idx')
```
